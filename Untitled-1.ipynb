```python
data = np.array([[0,0]])
di = 16
for x in range(di):
    data = np.append(data, [[x,x], [x, di-x]], axis=0)
    data = np.append(data, [[x,8], [8, x]], axis=0)
# data = np.append(data, [[2,3]], axis=0)
x, y = data.T
plt.scatter(x,y)
plt.show()
```
