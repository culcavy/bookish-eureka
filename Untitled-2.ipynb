```python
import math
```

```python
data = np.array([[0,0]])
di = 16
ang = 8
for x in range(math.floor(di)):
    for y in range(ang*2):
        data = np.append(data, [[np.clip(round(x*math.cos(math.pi/ang*y))+8, 0, di),np.clip(round(x*math.sin(math.pi/ang*y))+8, 0, di)]], axis=0)
    # data = np.append(data, [[x,x], [x, di-x]], axis=0)
    # data = np.append(data, [[x,8], [8, x]], axis=0)
# data = np.append(data, [[2,3]], axis=0)
data = np.unique(data, axis=0)
x, y = data.T
plt.scatter(x,y)
plt.show()
```
